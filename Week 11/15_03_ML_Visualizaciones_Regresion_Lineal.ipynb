### **CC57 - MACHINE LEARNING - VISUALIZACIONES E INTERPRETACION - MODELO REGRESION LINEAL**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos del ejemplo
horas_estudio = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
calificaciones = np.array([70, 75, 80, 85, 90])

# Crear el modelo de regresi√≥n lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir los valores ajustados
calificaciones_pred = modelo.predict(horas_estudio)

# Calcular residuos
residuos = calificaciones - calificaciones_pred


```

###**Visualizaciones**

```python
# Gr√°fica de dispersi√≥n con l√≠nea de regresi√≥n
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos Reales')
plt.plot(horas_estudio, calificaciones_pred, color='red', label='L√≠nea de Regresi√≥n')
plt.title('Gr√°fica de Dispersi√≥n con L√≠nea de Regresi√≥n')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaci√≥n')
plt.legend()
plt.grid(True)

# Gr√°fica de residuos
plt.subplot(1, 2, 2)
plt.scatter(horas_estudio, residuos, color='purple', label='Residuos')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Gr√°fica de Residuos')
plt.xlabel('Horas de Estudio')
plt.ylabel('Residuos')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### **Resultados Esperados e Interpretaci√≥n**

####**1. Gr√°fica de Dispersi√≥n con L√≠nea de Regresi√≥n**

- Muestra los puntos de datos y c√≥mo la l√≠nea de regresi√≥n se ajusta a estos puntos.

- Un buen ajuste se observa cuando la l√≠nea de regresi√≥n pasa cerca de la mayor√≠a de los puntos.

#### **2. Gr√°fica de Residuos**

- Muestra los residuos (diferencias entre valores observados y predichos) en funci√≥n de la variable independiente (horas de estudio).

- Idealmente, los residuos deber√≠an estar dispersos aleatoriamente alrededor de la l√≠nea horizontal en cero.

- La ausencia de patrones claros indica que el modelo es adecuado.

- En el caso de los datos proporcionados, esperamos ver una distribuci√≥n aleatoria de los residuos, confirmando que el modelo lineal es apropiado para estos datos.
