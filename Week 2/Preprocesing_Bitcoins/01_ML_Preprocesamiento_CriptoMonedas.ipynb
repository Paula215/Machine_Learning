```python
import pandas as pd
import os
```

A) Carga de datos

```python
metadata_df=pd.read_csv('metadata.csv')
metadata_df.head()
```

```python
# Diccionario para almacenar DataFrames
crypto_dataframes = {}

# Leer cada archivo CSV y almacenarlo en el diccionario
for index, row in metadata_df.iterrows():
    file_path = os.path.join('data', row['File Path'])  # Suponiendo que los archivos CSV estÃ¡n en la carpeta 'data'
    if os.path.exists(file_path):
        crypto_dataframes[row['Coin Pair Symbol']] = pd.read_csv(file_path)
        print(f"Datos cargados desde {file_path}")
    else:
        print(f"Archivo no encontrado: {file_path}")
```

```python
crypto_dataframes
```

B) Realizar tareas de prepocesamiento sobre los datos

```python
# Combinar todos los DataFrames en uno solo
combined_df = pd.concat(crypto_dataframes.values(), keys=crypto_dataframes.keys(), names=['Symbol', 'Index'])

# Resetear el Ã­ndice para un acceso mÃ¡s fÃ¡cil
combined_df.reset_index(level='Symbol', inplace=True)

# Mostrar las primeras filas del DataFrame combinado
print(combined_df.head())
```
